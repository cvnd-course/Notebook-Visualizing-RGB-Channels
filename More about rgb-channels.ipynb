```python
tela = cv2.imread('github_1713351.jpeg')
tela = cv2.cvtColor(tela,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,10))
plt.imshow(tela)
plt.title('Foto Rafaelszp')
plt.show()
```

```python
img = np.copy(tela)
print(img.shape)
img = img/255.
r = img[:,:,0]
print('red',r.shape)
rg = img[:,:,:-1]
print('red green',rg.shape)
g = img[:,:,1]
print('green',g.shape)
gb = img[:,:,1:]
print('green blue',gb.shape)
b = img[:,:,2]
print('blue',b.shape)
```

```python
#f, ((ax11,ax12),(ax21,ax22),(ax31,ax32)) = plt.subplots(2,3,figsize=(20,10))
f, ((ax1, ax2,ax3), (ax4, ax5,ax6)) = plt.subplots(2, 3, sharex='col', sharey='row')
f.suptitle('Red green blue in rgb channels and gray channel')

ax1.imshow(r,cmap='Reds')
#ax1.set_title('red channel')

ax2.imshow(g,cmap='Greens')
#ax2.set_title('green channel')

ax3.imshow(b,cmap='Blues')
#ax3.set_title('blue channel')

ax4.imshow(r,cmap='gray')
ax4.set_title('red channel')

ax5.imshow(g,cmap='gray')
ax5.set_title('green channel')

ax6.imshow(b,cmap='gray')
ax6.set_title('blue channel')
```

```python
sunglasses = cv2.imread('sunglasses.png', cv2.IMREAD_UNCHANGED)
plt.imshow(sunglasses)
plt.show()
```

```python
sunglasses.shape
```

```python
img_copy = np.copy(sunglasses)
no_alpha = img_copy[:,:,0:3]
alpha = img_copy[:,:,3]

f,(ax1,ax2) = plt.subplots(1,2)

ax1.imshow(no_alpha)
ax1.set_title('without alpha channel')

ax2.imshow(alpha,cmap='gray')
ax2.set_title('only alpha channel')

plt.show()
```
